# Limpieza de Base de Menciones de Xenofobia
Este notebook explicado sirve como guía para la limpieza de un archivo ```.xlsx``` extraído de Brandwatch que captura la conversación de migrantes venezolanos para un periodo determinado. En la primera versión del documento se usa el Piloto Barómetro 2021

````python
# Definimos el usuario que quiere hacer la limpieza (solo el primer nombre). Ej: Jose, Diego. Si usted no es alguno de los autores, escriba el PATH del archivo
user = input('Escriba su primer nombre - ')
if user == 'Jose':
    path = r"\Users\JOSE\Desktop\Trabajo\Base\mentions.xlsx"
elif user == 'Diego':
    path = r"/Users/diegoalejandrobermudezsierra/OneDrive - Universidad de los Andes/BAROMETRO DE XENOFOBIA/XENOFOBIA/mentions_xenofobia.xlsx"
else:
    path = input('Escriba la ruta del archivo - ')
#Importamos la base de datos que descargamos con los tweets de brandwatch 
xenofobia=pd.read_excel(path)
del path
xenofobia.head(11)
````

````python
#Eliminamos las primeras filas del df que tienen información que no necesitamos
#Eliminamos la primera columna que es un conteo dado por brandwatch
if user == 'Jose': numero = 6
elif user == 'Diego': numero = 7
else: numero = 7
xenofobia=xenofobia.iloc[numero:len(xenofobia),1:]
xenofobia.head(11)
````

````python
#Tomamos la primera fila como nombre de las columnas
column_names=xenofobia.iloc[0,:].tolist()
#Cambiamos los espacios por _
column_names=[re.sub('\\s', '_', x) for x in column_names]
#Renombramos las columnas
xenofobia.columns=column_names
#Eliminamos la fila que tomamos como nombre de las columnas
xenofobia=xenofobia.iloc[1:,:]
#Hacemos un reset al index
xenofobia=xenofobia.reset_index(drop=True)
del column_names
xenofobia.head()
````

````python
# Vamos a quedarnos solamente con los tweets que tienen tags de xenofobia. Primero, filtramos las columnas con tags de xenofobia
sub = "Xenofobia"
xenofobia["Contiene_Xenofobia"]= xenofobia["Categoría_Barometro_v4"].str.find(sub)
xenofobia[['Contiene_Xenofobia', 'Categoría_Barometro_v4']]
````

````python
# Se crea una nueva base con solo los tags de Xenofobia 
xenofobia_clean = xenofobia[xenofobia['Contiene_Xenofobia']<9.0]
print(xenofobia_clean[['Contiene_Xenofobia', 'Categoría_Barometro_v4']])
print('Se reportaron ' + str(len(xenofobia_clean)) + ' mensajes xenofobos')
````

````python
# Importamos autores de medios
text = open("medios.txt","r")
print(text)
medios = list()
for line in text:
    line = line.rsplit("OR")
    line = line[0].strip()
    if line.find(".") == -1:
        line = "@" + line
    medios.append(line)
medios
````

````python
# Vamos a guardar la lista de autores. Primero policia, luego medios tradicionales y después, alternos
medios_policia = ["@PoliciaBogota", "@region6policia", "@region7policia", "@Region8Policia", "@Region2Policia", "@Region3Policia", "@Region1Policia",
                    "@Region4Policia", "@Region5Policia", "@PoliciaNarino", "@DevalPolicia", "@PoliciaMeta", "@PoliciaMedellin", "@PoliciaBoyaca",
                    "@PoliciaMzales", "@PoliciaVillavo", "@PoliciaBmanga", "@PoliciaCali", "@PoliciaDEMAG", "@PoliciaCmarca", "@Policiantioquia",
                    "@PoliciaCucuta", "@PoliciaQuindio", "@Policia_Caldas", "@SucrePolicia", "@PoliciaVichada", "@PoliciaGuainia",
                    "@PoliciaArauca", "@PoliciaCasanare", "@GuajiraPolicia", "@PoliciaVaupes", "@PoliciaChoco", "@PoliciaDepuy",
                    "@PoliciaUraba", "@PoliciAtlantico", "@PoliciaDECOR", "@PoliciaNteSder", "@HuilaPolicia", "@PoliciaBolivar", "@PoliciaCaqueta",
                    "@PoliciaDeris", "@PoliciaDeTolima", "@PoliciaDEMAM", "@PoliciaMTunja", "@PoliciaMonteria", "@PoliciAmazonas", "@PoliciaDeCesar",
                    "@PoliciaSanyProv", "@PoliciaPasto", "@PoliciaStander", "@PoliciaStaMarta", "@PoliciaPereira", "@PoliciaIbague", "@PoliciaPopayan", 
                    "@PoliciaCtagena", "@PoliciaColombia"] 
medios_policia
````

````python
#Medios alternativos
text = open("alternos.txt","r")
print(text)
alternos = list()
for line in text:
    line = line.strip()
    alternos.append(line)
alternos
````

````python
#Búsqueda de tweets que citan a los autores
todos_medios = medios + medios_policia + alternos
def search_autor(df, autores):
    """
    Retorna una pequeña DataFrame en donde todas las filas tienen almenos una mención de medios o policía
    """
    return df[np.logical_or.reduce([df['Snippet'].str.contains(autor) for autor in autores])]

datos = search_autor(xenofobia_clean, tuple(todos_medios))
datos
````

````python
datos['Snippet']
````

````python
#Base consolidada
def label_base(df, autores):
    """
    Retorna una pequeña DataFrame en donde todas las filas tienen almenos una mención de medios o policía
    """
    df['Cita_medios'] = np.logical_or.reduce([df['Snippet'].str.contains(autor) for autor in autores])
    return df

xenofobia_clean = label_base(xenofobia_clean, tuple(todos_medios))
xenofobia_clean['Cita_medios'] = pd.get_dummies(xenofobia_clean['Cita_medios'])
xenofobia_clean['Cita_medios']
````

````python
#Tomamos solo la columna Full Text
tweets=pd.DataFrame(xenofobia.Full_Text)

#Limpiamos los tweets

#Eliminamos todos los @Usuario (i.e. @ClaudiaLopez)
tweets['Clean_Text']=tweets.Full_Text.replace('@\\w+', '', regex=True)
#Eliminamos todos los signos de puntuación
tweets['Clean_Text']=tweets.Clean_Text.replace(r'[^\w\s]+', '', regex=True)
#Remplazamos los acentos de las palabras y los emojis
tweets['Clean_Text']=tweets.Clean_Text.str.normalize('NFKD')\
        .str.encode('ascii', errors='ignore')\
        .str.decode('utf-8')
#Eliminamos los dobles espacios 
tweets['Clean_Text']=tweets.Clean_Text.replace('\s+', ' ', regex=True).str.strip()

#Exportamos la base limpia en formato txt
path="/Users/diegoalejandrobermudezsierra/OneDrive - Universidad de los Andes/BAROMETRO DE XENOFOBIA/XENOFOBIA/Xenofobia_Clean.txt"
tweets.Clean_Text.to_csv(path, header=None, index=None)

#Para efectos del ejercico de afinar la regla de Xenofobia dejamos los
#"Tweets unicos"
path="/Users/diegoalejandrobermudezsierra/OneDrive - Universidad de los Andes/BAROMETRO DE XENOFOBIA/XENOFOBIA/Xenofobia_Clean_Unicos.txt"
tweets_unicos=pd.DataFrame(tweets.Clean_Text.unique())
tweets_unicos.to_csv(path, header=None, index=None)

del path, tweets, tweets_unicos, xenofobia

````
